```python
# Provoa raffigurare tramite scatter plot le relazioni tra le features e il BEP
fig, axs = plt.subplots(5, 4)
fig.set_size_inches(18.5, 10.5)

for ax, feature in zip(fig.axes, df[:0]):
  ax.scatter(df.BEP,df[feature])
  ax.set_xlabel('bep')
  ax.set_ylabel(feature)

fig.tight_layout()
```
